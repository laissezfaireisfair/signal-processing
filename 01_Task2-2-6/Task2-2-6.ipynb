```kotlin
%use kandy
import kotlin.math.sin
```

$$\sin x \approx \sum_{i=1}^n -1\cdot \frac {x^{2i+1}} {(2i + 1)!}$$

```kotlin
fun approxSin(x: Double, termsCount: Int) = (1..<termsCount).fold(x to x) { (sum, lastTerm), i ->
    (lastTerm * -1 * x * x / 2 * i / (2 * i + 1)).let { newTerm -> sum + newTerm to newTerm }
}.first

val termsCountsToTest = listOf(1, 3, 7)

val args = generateSequence(-PI / 2) { if (it < PI / 2) it + 0.01 else null }.toList()
```

```kotlin
val lines = termsCountsToTest
    .map { termsCount -> "$termsCount" to args.map { approxSin(it, termsCount) } }
    .run {
        plus("reference" to args.map { sin(it) })
            .plus("args" to args)
    }
    .toMap()
```

```kotlin
plot(lines) {
    line {
        x("args")
        y("reference")
        color = Color.RED
    }
    termsCountsToTest.mapIndexed { i, terms ->
        line {
            x("args")
            y("$terms")
            color = Color.rgb(255, 255, 255 - i * 100)
        }
    }
}
```
